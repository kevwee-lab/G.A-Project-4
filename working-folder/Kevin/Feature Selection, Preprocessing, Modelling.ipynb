```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle


from sklearn.preprocessing import StandardScaler, PolynomialFeatures
#from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, plot_roc_curve, precision_recall_fscore_support,  roc_auc_score
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.feature_selection import RFE
from sklearn import metrics
from xgboost import  XGBClassifier
```

```python
comb = pd.read_csv('../../data/combined_kev.csv')
```

```python
comb.head()
```

```python
# convert sunrise & sunset to datetime
sunrise = pd.to_datetime(comb['sunrise'], format='%H%M')
# fix entries - round off 60min to the next hour
sunset = pd.to_datetime(comb['sunset'].map(lambda x: 1800 if x == 1760 else x), format='%H%M')
# convert to seconds from start of day to be consistent input into model
comb['sunrise'] = (sunrise - sunrise.dt.normalize()).dt.seconds
comb['sunset'] = (sunset - sunset.dt.normalize()).dt.seconds
# add total sunlight time feature
comb['total_sunlight_time'] = (sunset - sunrise).dt.seconds
```

```python
comb.head()
```

### Accuracy vs Interpretability

We see that most of the features we want to use for modelling are not within our control. This means that the measures to combat the west nile virus are not inferred from our modelling, therefore our main goal is to develop a model that optimizes accuracy to predict the west nile virus, rather than interpretability.

### Feature selection and Preprocessing Workflow

As we have decided to optimise for accuracy (ROC-AUC) instead of opting for more interpretability, we will use: 

**Preprocessing**
1) Polynomial features (degree of 2) to increase the number of features. <br>
2) Standard Scaler

**Feature Selection**
1) Pearson's correlation with 3 different cutoff points for correlation (> 0.01, > 0.05 and > 0.1) <br>
2) Principal Component Analysis <br>

**Modelling** <br>
We will find the best parameters using GridSearch on these models:
1) Support Vector Classifier: Effective in high dimensional spaces. Uses a subset of training points in the decision function.<br>
2) Logistic Regression: Effective in binary classification. <br>
3) Gradient boost: Produces a prediction model in the form of an ensemble of weak prediction models, typically decision trees. Uses gradient descent algorithm.<br>
4) Xg boost: Similar to gradient boost, but xgboost uses a more regularized model formalization to control over-fitting.<br>
5) Random Forest: Constructs decision trees at training time and outputting the class that is the mode of the classes.<br> 
6) Extra Trees: Similar to Random Forest. However, the splits of the trees in the Random Forest are deterministic. It is random for extratrees. <br>

We will then further fine tune the parameters, and select the best model based on ROC-AUC score, precision and recall.

```python
X = comb.drop(columns=['date', 'species', 'trap', 'wnvpresent'])
y = comb['wnvpresent']
poly = PolynomialFeatures(include_bias=False, degree=2)
X_poly = poly.fit_transform(X)
X_poly = pd.DataFrame(X_poly, columns=poly.get_feature_names(X.columns))
corr_df = X_poly.join(y).corr()[['wnvpresent']].abs().sort_values(by='wnvpresent', ascending=False)

pd.set_option('display.max_rows', len(corr_df))

# feature list
corr_df[corr_df['wnvpresent'] > 0.01]
```

```python
feature_list = list(corr_df[corr_df['wnvpresent'] > 0.01].index)

feature_list.remove('wnvpresent')

len(feature_list)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_poly[feature_list],
                                                    y,
                                                    test_size=0.3,
                                                    stratify=y,
                                                    random_state=42)
```

```python
ss = StandardScaler()
X_train_sc= ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)
```

### XG Boost

XGBoost is an implementation of gradient boosted decision trees designed for speed and performance. It is said to work well on structured or tabular datasets on classification and regression predictive modeling problems. 

Other than having better control on overfitting, xgboost is also said to have the following advantages over other boosted models:

1) Use of sparse matrices with sparsity aware algorithms <br>
2) Improved data structures for better processor cache utilization which makes it faster <br>
3) Better support for multicore processing which reduces overall training time <br>

```python
from xgboost import XGBClassifier
```

```python
xgb_model = XGBClassifier(use_label_encoder=False, objective='binary:logistic', eval_metric='auc')
```

```python
cross_val_score(xgb_model, X_train_sc, y_train, cv=5, scoring='roc_auc').mean()
```

```python
#Gridsearching for best params

parameters = {
    'max_depth': [3, 4, 5],
    'gamma' : [ 0.0, 0.2 , 0.4 ],
    'eta' : [0.05, 0.15, 0.25],
}

gs_xgb_model = GridSearchCV(
    xgb_model,
    param_grid = parameters,
    scoring='roc_auc',
    cv = 5)

gs_xgb_model.fit(X_train_sc, y_train)

print("Best parameter (CV score=%0.3f):" % gs_xgb_model.best_score_)
print(gs_xgb_model.best_params_)
```

```python
 # Scoring metrics
    
scores_xgb_1 = {'Classifier': 'xgboost', 'Cutoff': 0.01}
y_preds = gs_xgb_model.predict(X_test_sc)
y_pred_probas = gs_xgb_model.predict_proba(X_test_sc)[:, 1]
y_train_pred_probas = gs_xgb_model.predict_proba(X_train_sc)[:, 1]
scores_xgb_1['Train ROC-AUC Score'] = metrics.roc_auc_score(y_train, y_train_pred_probas)
scores_xgb_1['Test ROC-AUC Score'] = metrics.roc_auc_score(y_test, y_pred_probas)
scores_xgb_1['F1'] = metrics.f1_score(y_test, y_preds)
scores_xgb_1['Precision'] = metrics.precision_score(y_test, y_preds)
scores_xgb_1['Recall'] = metrics.recall_score(y_test, y_preds)
scores_xgb_1['Accuracy'] = metrics.accuracy_score(y_test, y_preds)
```

```python
scores_xgb_1
```

```python
#calculate roc curve

fig, ax = plt.subplots(figsize=(15,6))
plot_roc_curve(gs_xgb_model, X_test_sc, y_test, ax=ax)
```

```python
# Save the model to file 

Pkl_Filename = "xgb_model.pkl"  

with open(Pkl_Filename, 'wb') as file:  
    pickle.dump(gs_xgb_model.best_estimator_, file)
```

Our first xgboost model seems to give us pretty decent results. However, it is to be noted that the recall and f1 score is low. This is possibly due to the imbalance in classes, which results in the model not being able to provide us with more true positives.

### ExtraTrees

Extremely Randomized Trees, or Extra Trees for short, is an ensemble machine learning algorithm.

Specifically, it is an ensemble of decision trees and is related to other ensembles of decision trees algorithms such as bootstrap aggregation (bagging) and random forest.

The Extra Trees algorithm works by creating a large number of unpruned decision trees from the training dataset. Predictions are made by averaging the prediction of the decision trees in the case of regression or using majority voting in the case of classification.

Unlike random forest, which uses a greedy algorithm to select an optimal split point, the Extra Trees algorithm selects a split point at random.

```python
et_model = ExtraTreesClassifier()
```

### Finetuning XGBoost by expanding parameters

We know that xgboost is by far the best model, after evaluating the table above. However, are we able to further optimise the ROC AUC score by finetuning the model? In this section we attempt to do so by adding in more parameters for max_depth, gamma and eta. We also add subsample and set it to 0.5. Setting the subsample to 0.5 means that XGBoost would randomly sample half of the training data prior to growing trees. and this will prevent overfitting. Subsampling will occur once in every boosting iteration.

```python
xgb_model_x = XGBClassifier(use_label_encoder=False, objective='binary:logistic', eval_metric='auc')
```

```python
#Gridsearching for best params

parameters = {
    'max_depth': [3, 4, 5, 6],
    'gamma' : [0.0, 0.2, 0.3, 0.4],
    'eta' : [0.05, 0.15, 0.25, 0.4],
    'subsample': [0.5, 1]
}

gs_xgb_model_x = GridSearchCV(
    xgb_model_x,
    param_grid = parameters,
    scoring='roc_auc',
    cv = 5)

gs_xgb_model_x.fit(X_train_sc, y_train)

print("Best parameter (CV score=%0.3f):" % gs_xgb_model_x.best_score_)
print(gs_xgb_model_x.best_params_)
```

```python
scores_xgb_x = {'Classifier': 'xgb_model_x', 'Cutoff': 0.01}
y_preds = gs_xgb_model_x.predict(X_test_sc)
y_pred_probas = gs_xgb_model_x.predict_proba(X_test_sc)[:, 1]
y_train_pred_probas = gs_xgb_model_x.predict_proba(X_train_sc)[:, 1]
scores_xgb_x ['Train ROC-AUC Score'] = metrics.roc_auc_score(y_train, y_train_pred_probas)
scores_xgb_x ['Test ROC-AUC Score'] = metrics.roc_auc_score(y_test, y_pred_probas)
scores_xgb_x ['F1'] = metrics.f1_score(y_test, y_preds)
scores_xgb_x ['Precision'] = metrics.precision_score(y_test, y_preds)
scores_xgb_x ['Recall'] = metrics.recall_score(y_test, y_preds)
scores_xgb_x ['Accuracy'] = metrics.accuracy_score(y_test, y_preds)
```
