```python
test = pd.read_csv('../../data/test.csv')
weather = pd.read_csv('../../data/weather_eng.csv', index_col=0)
```

```python
weather.head()
```

```python
test.head()
```

```python
#lowercase columns
test.columns = [x.lower() for x in test.columns]
```

```python
test['species_ord'] = test['species'].map({'CULEX PIPIENS/RESTUANS': 2, 'CULEX PIPIENS': 2, 'CULEX RESTUANS': 1})

#0 for species other than the virus carriers
test['species_ord'].fillna(0, inplace=True)
```

```python
test.head()
```

```python
#add week number

test['date'] = pd.to_datetime(test['date'], format='%Y-%m-%d')

test['week_number'] = test['date'].dt.week
```

```python
#drop address related features
test.drop(['address', 'block', 'street', 'addressnumberandstreet', 'addressaccuracy'], axis=1, inplace=True)
test.head()
```

```python
weather['date'] = pd.to_datetime(weather['date'], format='%Y-%m-%d')
```

```python
#merge dataframes

test_1 = test.merge(weather, on='date')
test_1.head()
```

```python
# convert sunrise & sunset to datetime
sunrise = pd.to_datetime(test_1['sunrise'], format='%H%M')
# fix entries - round off 60min to the next hour
sunset = pd.to_datetime(test_1['sunset'].map(lambda x: 1800 if x == 1760 else x), format='%H%M')
# convert to seconds from start of day to be consistent input into model
test_1['sunrise'] = (sunrise - sunrise.dt.normalize()).dt.seconds
test_1['sunset'] = (sunset - sunset.dt.normalize()).dt.seconds
# add total sunlight time feature
test_1['total_sunlight_time'] = (sunset - sunrise).dt.seconds
```

```python
#drop date, trap, species
test_1.drop(columns=['date', 'trap', 'species'], inplace=True)

test_1.head()
```

```python
test_1.to_csv('../../data/prepoly_combined_test.csv')
```
